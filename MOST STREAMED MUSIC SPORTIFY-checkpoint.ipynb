```python
# checking for missing values and filling those value

def fill_missing(df, column):
    if df[column].dtype in ['int64', 'float64']:
        return df[column].fillna(df[column].median())
    else:
        return df[column].fillna(df[column].mode() [0])
```

```python
column_with_missing = ['Artist', 'Spotify Streams', 'Spotify Playlist Count', 'Spotify Playlist Reach', 'Spotify Popularity',
                       'YouTube Likes', 'YouTube Views', 'TikTok Posts', 'TikTok Likes', 'TikTok Views', 'YouTube Playlist Reach',
                       'Apple Music Playlist Count']
```

```python
for column in column_with_missing:
    df[column] = fill_missing(df, column)
```

```python
# Converting other columns to numeric

def convert_to_numeric(column):
    if column.dtypes == 'object':
        column = column.str.replace(',','', regex=False)
        return pd.to_numeric(column, errors = 'coerce')
```

```python
# Data Analysis
# Top 10 songs

top_10_songs = df.nlargest(10, 'Spotify Streams')
print("\nTop 10 Most Streamed Songs:")
print(top_10_songs[['Track', 'Artist', 'Spotify Streams']])
    
```

```python
# Plotting top 10 songs

top_10_songs = top_10_songs.sort_values('Spotify Streams', ascending=False)
plt.figure(figsize=(10,6))
plt.barh(top_10_songs['Track'], top_10_songs['Spotify Streams'], color='green')
plt.xlabel('Spotify Streams', fontsize=12)
plt.ylabel('Tracks', fontsize=12)
plt.title('Top 10 Most Stream Songs', fontsize=14)
plt.savefig('top10streamssongs.png')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Artists
top_10_artists = df.groupby("Artist")['Spotify Streams'].sum().nlargest(10)
```

```python
# Plotting top 10 Artist

top_10_artists.sort_values(ascending=False, inplace=True)
plt.figure(figsize=(12,6))
top_10_artists.plot(kind='bar')
plt.title('Top Artists by Total Spotify Streams')
plt.xlabel('Artist')
plt.ylabel('Total Spotify Streams')
plt.xticks(rotation=45,ha='right')
plt.xlim(-0.5, 9.5)
plt.savefig('topartistbystreams.png')
plt.tight_layout()
plt.show()
```

```python
# Getting to known the relatinship between streamingservices

correlation_columns  = ['Spotify Streams', 'Spotify Playlist Count', 'Spotify Popularity', 'YouTube Views', 'TikTok Posts']
correlation_matrix = df[correlation_columns].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", linewidths=.5,  cmap="viridis", annot_kws={"size":8})
plt.title('Correlation Heatmap of Music Matrics')
plt.xlabel('Metics')
plt.ylabel('Metrics')
plt.savefig('heatmap.png')
plt.tight_layout()
plt.show()
```

```python
# Spotify popularity distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['Spotify Popularity'], bins=50, kde=True, log_scale=True, color='green')
plt.title('Distribution of Spotify Popularity', fontsize=14, fontweight='bold')
plt.xlabel('Spotify Popularity', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', alpha=0.7)
plt.savefig('spotifypopularity.png')
plt.tight_layout()
plt.show()
```

```python
# Relationship between spotify and youtube views

plt.figure(figsize=(10, 8))
plt.scatter(df['Spotify Streams'], df['YouTube Views'], alpha=0.5, s=20)
plt.title('Relationship between Spotify Streams and YouTube Views')
plt.xlabel('Spotify Streams')
plt.ylabel('YouTube Views')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.savefig('relationshipbetweenspotifyandyoutubeviews.png')
plt.tight_layout()
plt.show()
```
